```python
import numpy as np
import random
from datetime import datetime, timedelta

# Define random data generators
def random_email(name):
    domains = ["@gmail.com", "@yahoo.com", "@outlook.com", "@company.com"]
    return name.lower().replace(" ", ".") + random.choice(domains)

def random_phone():
    return "+91" + str(random.randint(6000000000, 9999999999))

def random_services():
    services = ["cleaning", "beauty", "plumbing", "electrical", "carpentry", "painting"]
    return random.choice(services)

def random_time_slot():
    slots = ["Morning", "Afternoon", "Evening", "Night"]
    return random.choice(slots)

def random_date():
    return (datetime.today() - timedelta(days=random.randint(0, 365))).strftime("%Y-%m-%d")

# Add new columns with random values
# df["user_id"] = df["code"]
df["email"] = df["name"].apply(random_email)
df["phone_number"] = df["name"].apply(lambda x: random_phone())
df["location_id"] = np.random.randint(1, 100, size=len(df))
df["preferred_services"] = df["name"].apply(lambda x: random_services())
df["preferred_time_slots"] = df["name"].apply(lambda x: random_time_slot())
df["booking_history_count"] = np.random.randint(1, 50, size=len(df))
df["cancellation_rate"] = np.round(np.random.uniform(0, 1, size=len(df)), 2)
df["average_rating_given"] = np.round(np.random.uniform(1, 5, size=len(df)), 1)
df["wallet_balance"] = np.round(np.random.uniform(0, 5000, size=len(df)), 2)
df["last_booking_date"] = df["name"].apply(lambda x: random_date())
df["subscription_status"] = np.random.choice([0, 1], size=len(df))
df["feedback_score"] = np.round(np.random.uniform(1, 10, size=len(df)), 1)

# Drop the old `code` column to avoid redundancy
# df.drop(columns=["code"], inplace=True)

# Show a preview of the updated dataset
df
```
