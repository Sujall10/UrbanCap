```python
used_slots = {}
```

```python
# Function to create a booking record
def create_booking(user, total_bookings):
    for _ in range(total_bookings):
#         provider_id = SP['provider_id']
        provider_id = random.choice(SP['provider_id'].unique())  # Select a single provider
        service_id = random.choice(SP['service_Id'].unique())  # Select a single service
        service_location = user['location_id']  # Select a single ID from available services
        service_location = user['location_id']

        # Generate booking and service dates
        booking_date = datetime.now() - timedelta(days=random.randint(10, 180))
        service_date = booking_date + timedelta(days=random.randint(1, 7))  # Service within a week

        # Ensure provider is not booked at the same time elsewhere
        provider_id = SP['provider_id'].iloc[0]  # Fetch provider_id safely

        provider_key = (provider_id, service_date)

        if provider_key in used_slots:
            continue  # Skip booking if provider is already scheduled at that time

        used_slots[provider_key] = True  # Mark provider as booked

        # Generate booking details
        booking_status = random.choice(["Completed", "Cancelled", "Confirmed"])
        payment_status = random.choice(["Paid", "Pending", "Refunded", "Failed"])
        total_amount = round(random.uniform(500, 5000), 2)
        discount_applied = round(random.uniform(0, total_amount * 0.2), 2)  # Random discount up to 20%
        tax_amount = round(total_amount * 0.18, 2)  # 18% GST or service tax
        final_payment = round(total_amount - discount_applied + tax_amount, 2)

        booking_history.append([len(booking_history) + 1, user['user_id'], provider_id, service_id, booking_date, service_date,
                                service_location, booking_status, payment_status, total_amount,
                                discount_applied, tax_amount, final_payment])
```

```python
user = pd.read_excel('Datas/userUrban.xlsx')
for _, user in user.iterrows():  # Ensure it's user_df, not user
    total_bookings = user['booking_history_count']  # Adjust this to match the actual column name
    create_booking(user, total_bookings)
```
