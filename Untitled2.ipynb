```python
time_slot_mapping = {
    'Morning': (6, 12),
    'Afternoon': (12, 17),
    'Evening': (17, 22),
    'Night': (22, 6)  # Wraps around midnight
}
```

```python
import numpy as np
# x['preferred_time_midpoint'] = x['preferred_time_slots'].map(lambda x: np.get(time_slot_mapping.get(x, (0, 24))))
x['preferred_time_range'] = x['preferred_time_slots'].map(lambda slot: time_slot_mapping.get(slot, (0, 24)))
```

```python
x
```

```python
from datetime import datetime
x["Service_time"] = pd.to_datetime(x["Service_Time"], errors="coerce")
x["Service_hour"] = x["Service_time"].dt.hour
```

```python
# x['time_slot_match'] = (x['Service_Hour'].between(x['preferred_time_midpoint'] - 1, x['preferred_time_midpoint'] + 1)).astype(int)
```

```python
x = x.drop(columns={'Service_time'})
```

```python
x
```

```python
x['preferred_time_start'] = x['preferred_time_range'].apply(lambda x: x[0])
x['preferred_time_end'] = x['preferred_time_range'].apply(lambda x: x[1])
x['time_slot_match'] = x.apply(lambda row: row['preferred_time_start'] <= row['Service_hour'] < row['preferred_time_end'], axis=1)
x.drop(columns={'preferred_time_start','preferred_time_end'},inplace = True)
```
